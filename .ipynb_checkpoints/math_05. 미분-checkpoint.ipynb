```python
def average(a,b):
    x = symbols('x')
    
    fx = 2*x**2 + 4*x + 7
    fb = fx.subs(x,b) 
    fa = fx.subs(x,a) 
    
    result = (fb-fa)/(b-a)
    return result
print(average(0,2))
```

### 순간변화율

```python
from sympy import Derivative, symbols
```

```python
# 평균변화율을 구할 수 있는 함수를 정의
x = symbols('x')
fx = 2*x**2 + 4*x +7
```

```python
# fprime라는 Derivative 클래스의 객체를 생성한 후
# subs() 메서드를 사용하여 x=3에서의 미분계수 f'(3)을 구한다.
fprime = Derivative(fx,x).doit() # x에 대해 미분
n = fprime.subs({x:3})
print(fprime)
print("fx에서 x = 3에서의 순간변화율(미분계수는)",n,"입니다.")
```
