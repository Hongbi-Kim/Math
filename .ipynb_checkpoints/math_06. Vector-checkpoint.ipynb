```python
from scipy import linalg
```

```python
s = np.array([2,3,1])
```

```python
# 단위 벡터
s / linalg.norm(s)
```

```python
# Norm
x = np.array([1,-3])
v_hat = x / linalg.norm(x)
print(v_hat)
```

```python
# Norm 2
x = np.array([-6,-8])
v_hat2 = x / (x**2).sum()**0.5
print(v_hat2)
```

```python
# 선형 독립, 선형 종속
# Rank
# matrix_rank()

x = np.array([[2,7],[5,1]])
np.linalg.matrix_rank(x)
```

```python
x1 = np.array([[2,4],[3,6]])
np.linalg.matrix_rank(x1)
```

```python
x2 = np.array([[1,2],[3,5]])
np.linalg.matrix_rank(x2)
```

```python
# 벡터의 덧셈
x = [2,3]
y = [3,1]

z = [i+j for i,j in zip(x,y)]
print(z)
```

```python
# 벡터의 뺄셈
z2 = [i-j for i,j in zip(x,y)]
print(z2)
```

```python
# 벡터의 곱셈(inner product, outer product)
x1 = [3,4]
c = 8
z3 = [c*i for i in x1]
print(z3)
```

```python
# 내적(inner product)
u = np.array([6,6])
v = np.array([12,0])
np.dot(u,v)
```
