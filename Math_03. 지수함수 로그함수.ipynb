# 지수함수

```python
import math as ma
```

```python
# 2의 5 제곱
pow = ma.pow(2,5)
print("pow 결과 : ", pow)
```

```python
# 2의 제곱근
sqrt = ma.sqrt(2)
print("sqrt 결과 : ", sqrt)
```

```python
# e의 2 제곱
exp = ma.exp(2)
print("exp 결과 : ", exp)
```
