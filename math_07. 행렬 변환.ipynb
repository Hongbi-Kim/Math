## 1. 역함수와 역변환
- 역행렬

```python
import numpy as np
```

```python
A = np.matrix([[1,0,0,0],[2,1,0,0],[3,0,1,0],[4,0,0,1]])
print(A)
print()
print(np.linalg.inv(A))
```

- 전치행렬(Transpose)

```python
a = np.arange(15).reshape(3,5)
print(a)
```

```python
np.transpose(a)
```

- 고유값, 고유벡터

```python
a = np.array([[5,-1],[-2,1]])
w, v = np.linalg.eig(a)
print(w) # r고유값
print(v) # 고유벡터
```

```python
np.linalg.eig(a)
```

```python
a = np.array([[1,3],[4,2]])
w, v = np.linalg.eig(a)
print(w) # r고유값
print(v) # 고유벡터
```
